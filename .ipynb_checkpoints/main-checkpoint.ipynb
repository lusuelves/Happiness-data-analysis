```python
import requests
import numpy as np
import pandas as pd
import re
happy_data = pd.read_csv('Happy2019.csv', engine = 'python')
happy_data
```

```python
#There are no null values and every column is useful for analysis
null_cols = happy_data.isnull().sum()
print("Number of columns with any null value: " , len(null_cols[null_cols > 0]))
```

```python
#I added a column with the code for each country to search the climate api
def country_code(country):
    try:
        country_data = requests.get(f"https://restcountries.eu/rest/v2/name/{country}?fullText=true").json()
        return[country['alpha3Code'] for country in country_data][0]
    except:
        return None
countries = happy_data['Country or region']
happy_data['Code'] = [country_code(country) for country in countries]
null_codes = happy_data[happy_data['Code'].isnull() == True]
null_codes['Country or region']
```

```python
#I changed the code null values to the real ones, the API could not find
real_code = ["GBR","USA", "TTO", "RKS", "KOR", "BOL", "CYP", "RUS", "MDA", "MKD", "VNM","COD","VEN","PSE","IRN","COD","SYR","TZA"]
countries = null_codes['Country or region'].tolist()
correction = zip(countries,real_code)
def replace_code():
    for (country, code) in correction:
        happy_data.loc[happy_data['Country or region'] == country, 'Code'] = code
    return happy_data.head()
replace_code()
```

```python
#New column with average anual temperature 2012 (could not find data for 2019)
def temperature(code):
    try:
        temp_resp =  requests.get(f"http://climatedataapi.worldbank.org/climateweb/rest/v1/country/cru/tas/year/{code}.json").json()
        return [temp['data'] for temp in temp_resp if temp['year'] == 2012][0]
    except: 
        return None
codes = happy_data['Code'].tolist()
temperatures = [temperature(code) for code in codes]
temperatures
```

```python
happy_data['Average Temperature'] = temperatures
happy_data.head()
```

```python
#Null values for average temperature, can we solve it?
happy_data[happy_data['Average Temperature'].isnull() == True]
```

```python
#New column with average anual precipitations 2012 (could not find data for 2019)
def precipitations(code):
    try:
        temp_resp =  requests.get(f"http://climatedataapi.worldbank.org/climateweb/rest/v1/country/cru/pr/year/{code}.json").json()
        return [temp['data'] for temp in temp_resp if temp['year'] == 2012][0]
    except: 
        return None
codes = happy_data['Code'].tolist()
precipitations = [precipitations(code) for code in codes]
precipitations
```

```python
happy_data['Average Precipitations'] = precipitations
happy_data.head()
```

```python
#Null values for average temperature, can we solve it?
happy_data[happy_data['Average Precipitations'].isnull() == True]
```

```python
from bs4 import BeautifulSoup
import requests

res = requests.get("https://en.wikipedia.org/wiki/List_of_countries_by_suicide_rate")
soup = BeautifulSoup(res.text,'html.parser')
table = soup.find_all('tbody')[3]
rows = table.find_all('tr')
def suicide_rate(row):
    try:
        info = row.find_all('td')
        return {
            'Country': info[1].select('a[href][title]')[0].text,
            'Continent': info[2].text.strip(),
            'Suicide per 1000': info[3].text.strip()
        }
    except:
        return None
suicides = [suicide_rate(row) for row in rows][2:]
suicides = pd.DataFrame(suicides)
suicides
```

```python
#Data scraping from wikipedia to extract suicide rate, new column to de dataframe
import pandas as pd
happy_data['Country or region'].tolist()
happy_data['Suicide per 1000'] = 0
happy_data['Continent'] = ''
def what(country):
    try: 
        suicide = suicides.loc[suicides['Country'] == country]['Suicide per 1000'].iloc[0]
        continent = suicides.loc[suicides['Country'] == country]['Continent'].iloc[0]
        happy_data.loc[happy_data['Country or region'] == country, 'Suicide per 1000'] = suicide
        happy_data.loc[happy_data['Country or region'] == country, 'Continent'] = continent
        return happy_data.loc[happy_data['Country or region'] == country]
    except:
        happy_data.loc[happy_data['Country or region'] == country, 'Suicide per 1000'] = None
        return None
suicides_rate = [what(country) for country in happy_data['Country or region'].tolist()]
happy_data
```

```python
#Using selenium no get medals of olimpics 2016
from selenium import webdriver

class PageGetter:
    driver = False
    def __init__(self, defaultBrowser='firefox'):
        if defaultBrowser == 'firefox':
            self.driver = webdriver.Firefox()
        elif defaultBrowser == 'chrome':
            self.driver = webdriver.Chrome("/Users/luciasuelves/Downloads/chromedriver") 
        else:
            raise Exception('Not a browser')

    def getPage(self, url):
        if not self.driver:
            raise Exception("You should start a browser connection")
        self.driver.get(url)
        html = self.driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        return soup

    def close(self):
        self.driver.quit()
        
pg = PageGetter("chrome")
hdr_data = pg.getPage("https://www.bbc.com/sport/olympics/rio-2016/medals/countries")
pg.close()
#hdr_data.find_all('tr')
```

```python
medals = hdr_data.find_all('table')[0].find_all('tr')[1:]
def find(medal):
    try:
        return {
            "Country": medal.find('abbr')['title'],
            "Medals": medal.select('td.medals-table__total')[0].text.strip()
        }
    except: 
        return None
table = [find(medal) for medal in medals]
table = pd.DataFrame(table)
```

```python
import pandas as pd
happy_data['Olympic medals'] = 0
def medals(country):
    try: 
        medal = table.loc[table['Country'] == country]['Medals'].iloc[0]
        happy_data.loc[happy_data['Country or region'] == country, 'Olympic medals'] = medal
        return happy_data.loc[happy_data['Country or region'] == country]
    except:
        happy_data.loc[happy_data['Country or region'] == country, 'Suicide per 1000'] = None
        return None
result = [medals(country) for country in happy_data['Country or region'].tolist()]
```
